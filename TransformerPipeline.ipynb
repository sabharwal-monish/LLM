```python
import transformers
```

```python
from transformers import pipeline
```

```python

classifier = pipeline("sentiment-analysis")
classifier("I've been waiting for a HuggingFace course my whole life.")
```

```python
generator = pipeline('text-generation')
generator("In this course, we will teach you how to", max_length = 15, num_return_sequences= 3)
```

```python
generator = pipeline('text-generation', model = 'HuggingFaceTB/SmolLM2-360M')
generator("In this course, we will teach you how to", max_length = 15, num_return_sequences= 2)
```

```python
from transformers import pipeline

translator = pipeline("translation", model="Helsinki-NLP/opus-mt-fr-en")
translator("Ce cours est produit par Hugging Face.")
```

```python
from transformers import pipeline
from google.colab import files
from PIL import Image


uploaded = files.upload()

image_path = list(uploaded.keys())[0]
print(f"Uploaded file: {image_path}")

from IPython.display import Image, display

display(Image(filename=image_path))

image_classifier = pipeline(
    task="image-classification",
    model="google/vit-base-patch16-224"
)
result = image_classifier(image_path)

print(result)
```
